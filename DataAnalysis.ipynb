```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_info_columns', 1000)
df = pd.read_csv("data.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# removing irrelevant columns
df=df[['sofifa_id', 'short_name', 'age', 'dob',
       'height_cm', 'weight_kg', 'nationality', 'club_name',
       'league_name', 'league_rank', 'overall', 'potential', 'skill_moves', 'pace', 'shooting', 'passing', 'dribbling',
       'defending', 'physic', 'player_traits',
       'attacking_crossing', 'attacking_finishing',
       'attacking_heading_accuracy', 'attacking_short_passing',
       'attacking_volleys', 'skill_dribbling', 'skill_curve',
       'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control',
       'movement_acceleration', 'movement_sprint_speed',
       'movement_agility', 'power_stamina',
       'power_strength', 'mentality_aggression',
       'mentality_vision', 'mentality_composure',]]
df
```

```python
df.info(verbose = True)
```

```python
# Dropping rows with missing values
df.replace("", np.nan, inplace=True)
df.dropna(axis=0, inplace=True)
```

```python
# Converting and height and weight columns to float
df['height_cm']=df['height_cm'].astype(float)
df['weight_kg']=df['weight_kg'].astype(float)
df
```

```python
# Plotting a regression plot of mentality_vision against a player's overall performance
sns.regplot(x=df["mentality_vision"], y=df["overall"], line_kws={"color": "purple"})
```

```python
# Plotting a regression plot of dribbling against a player's overall performance
sns.regplot(x=df["dribbling"], y=df["overall"], line_kws={"color": "purple"})
```

```python
# Plotting a Histogram to show distribution of the players' ages using matplotlib
plt.figure(figsize=(8,8))
n, bins, edges= plt.hist(df['age'], bins=10, edgecolor="black")
plt.xticks(np.round(bins,0))
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Players' Ages")
plt.show()
```

```python
# Plotting a scatter plot to show how the player's weights and heights are correlated using matplotlib
plt.figure(figsize=(8,8))
plt.scatter(x=df['weight_kg'], y=df['height_cm'])
plt.xlabel('Weight in Kg')
plt.ylabel('Height in cm')
plt.title('Scatter Plot of Weight Against Height')
plt.show()
```
